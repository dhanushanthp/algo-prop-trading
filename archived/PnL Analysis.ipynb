```python
import plotly.express as px
import pandas as pd

# Sample data
# Break
df = pd.read_csv("10_1_0_8.csv", names = ["Time", "strategy", "trail", "profit_factor", "rr","pnl"])

# Convert the 'Time' column to datetime type
df['Time'] = pd.to_datetime(df['Time'])

# Create the plot
fig = px.line(df, x='Time', y='pnl', title='Time vs Number Plot')

# Show the plot
fig.show()
```

```python
# Create the plot
fig = px.line(df, x='Time', y='rr', color='trail', title='Time vs Number Plot')

# Show the plot
fig.show()
```
